```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic["age"].fillna(titanic["age"].mean(), inplace = True)
bins = [1,20,30,50,70,100]
labels = ["미성년자","청년","장년","중년","노년"]
titanic
```

```python
titanic[(titanic["age"] <1) | (titanic["age"] >100)]
```

```python
titanic_age = pd.cut(titanic["age"],bins=bins, labels=labels).value_counts()
```

```python
titanic_age_rate = titanic_age / titanic_age.sum()

df_titanic_age = pd.DataFrame({"연령층" : titanic_age, "연령비율": titanic_age_rate}, index = labels)
df_titanic_age.loc["합계"] = df_titanic_age.sum(axis = 0)
df_titanic_age.astype({"연령층":"int32"})
```

```python
titanic["연령층"] = pd.cut(titanic["age"], bins=bins, labels=labels)
titanic["category3"] = titanic.apply(lambda i : "미성년자" if i.age < 20 else i["연령층"]
                                     +("남성" if i ["sex"] == "male" else "여성"), axis=1)
titanic[["연령층","age","sex","category3"]]
```

```python
def cate3(a):
    bins = [20,30,50,70,100]
    labels = ["청년","장년","중년","노년"]
    if a.age < 20:
        return "미성년자"
    elif a.age >= 20:
        return a.sex
    return ""
titanic["category3"] = titanic.apply(cate3, axis=1)
wwwtitanic[["category3"]]
```

```python
employees = pd.read_csv(
    "datas/employees.csv", parse_dates = ["Start Date"]
)
```

```python
employees.dropna()
```

## 행 안의 모든 값이 결측치를 가지면 how매개변수로 

```python
employees.dropna(how = "all")
```

```python
employees.loc[employees.Gender.isnull()]
```

```python
#특정 컬럼에 대해서만 NaN값이 있는 행을 지우는 법
#subset키워드 인수로 컬럼 인덱스 전달
employees.dropna(subset = ["Gender"])
```

```python
# 두 조건에 대해 모두 만족하는 조건
cond1 = employees["Start Date"].isnull()
cond2 = employees["Salary"].isnull()
```

```python
# 두 조건에 대해 모두 만족하는 조건
employees[cond1 | cond2]
```

```python
#두 조건에 대해 모두 만족하는 조건
# subset키워드 인수에 리스트의 형태로 컬럼 인덱스인 ["Start Date","Salary"]를 전달
# 3개 행 데이터 제거
employees.dropna(subset = ["Start Date","Salary"])
```

```python
employees[employees.notnull().sum(axis=1) >= 4]
```

```python
#행중에 최소 n개 이상 값이 존재 하는 행을 가져온다, -> thresh인수
# thresh = 4 >>행에 최소 4개 이상 값이 이써어야 보존 없으면 제거
employees.dropna(how = "any", thresh = 4)
```

# duplicated() .중복처리하기

## 하나의 컬럼에 대한 중복 값이 존재하는지 Series객체에서의 중복 확인 방법

```python
# Series객체를 순회 하면서 한번이라도 본 적이 있는 값에 대해 True반환
```

```python
employees["Team"].head()
```

```python
employees["Team"].duplicated().head()
```

```python
# keep 매개변수에  * last전달 :(마지막 발견된 값 -> False 반환) (앞에 발견 -> True)
```

```python
employees["Team"]
```

```python
employees["Team"].duplicated(keep = "last")
```

## duplicated() 메서드 결과를 반전 시키면 첫번째로 등장한 값만 True 반환

```python
# ~ 기호는 반전 bool type
```

```python
first_one_in_team = ~employees["Team"].duplicated()
employees[first_one_in_team]
```

```python
# drop_duplicates() 인수를 사용하지 않으면 모든 값이 일치하는 행 제거
```

```python
employees.drop_duplicates()
```

```python
#특정 컬럼 중복 제거 subset= 컬럼
```

```python
employees.drop_duplicates(subset = ["Team"])
```

```python
# 마지막에 발견된 값만 남기고 이전에 나오는 중복 제거 keep = 'last'
```

```python
employees.drop_duplicates(subset = ["Team"], keep = 'last')
```

```python
# Keep=False -> 중복이 있는 모든 값 제거
```

```python
employees.drop_duplicates(subset = ["First Name"], keep = False)
```
