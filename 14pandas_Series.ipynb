```python
df_tm = pd.read_csv("datas/teams.csv").squeeze()

data = {}
team_num = 5
for i in range(math.ceil(len(df_tm)/team_num)):
    if len(df_tm) < team_num:
        _team = df_tm.sample(len(df_tm))
    else:
        _team = df_tm.sample(team_num)
    data[str(i + 1) + "팀"] = [_team.values]
    df_tm.drop(axis=0, index=_team.index, inplace=True)
```
