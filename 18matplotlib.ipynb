## 히스토그램 - displot() 2개 값을 kde 표현

```python
# x,y축 모두 할당한 앞의 히스토그램을 kde로 표현 가능
sns.displot(penguins,
            x="flipper_length_mm",
            y="bill_length_mm",
            kind="kde",
            rug=True
)
```

## 히스토그램 - displot() col 나누기

```python
# displot()은 col 키워드 인수로 subset 한번더 분류 가능
sns.displot(penguins,
            x="flipper_length_mm",
            hue="species",
            col="sex",
            kind = "kde"
)
```

## 히스토그램 - displot() 그래프 크기 설정

```python
# hight와 aspect를 활용해서 그래프의 크기를 제어 할 수 있다.
sns.displot(penguins,
            y="flipper_length_mm",
            hue="sex",
            col="species",
            kind = "ecdf",
            height=4,
            aspect=.7
)
```

## 히스토그램 - displot() col로 그래프 나누기

```python
# axis label / title 지정 가능.
# axis label = set_axis_labels(xlabel.ylabel)
# title = set_titles() -> {col_var} / {col_name}
g = sns.displot(penguins,
                y="flipper_length_mm",
                hue="sex",
                col="species",
                kind = "kde",
                height=4,
                aspect=.7
)
g.set_axis_labels("Density (a.u.)", "Flipper length (mm)")
g.set_titles("{col_name} penguins")
```

# violinplot()

# stripplot()
## -> 다른 축의 방향으로 랜덤하게 점을 찍어 겹피는 부분을 줄인다

```python
# 단 하나의 숫자형column을 할당 
#-> 각 점들이 다른축의 방향으로 랜덤하게 "jittered"단일 변량에 대한 분포
tips=sns.load_dataset("tips")
sns.stripplot(data=tips, x="total_bill")
```

# swarmplot()
## -> 절대로 겹치지 않게 표현 (아주 큰 데이터를 표현하기엔 적합하지 않음)

```python
# 숫자값을 갖는 단일 변량의 분포 표현
tips = sns.load_dataset("tips")
sns.swarmplot(data=tips, x="total_bill")
```

```python
# y값에 카테고리 자료형을 갖는 column을 추가
#요일별 지출액에 대한 분포를 나타냄
sns.swarmplot(data=tips, x="total_bill",y="day")
```

```python
# x값에 카테고리 자료형을 y값에 숫자형 데이터 설정 - 축 반대
sns.swarmplot(data=tips, y="total_bill",x ="day")
```

```python
# hue키워드 인수 추가 - 변량 추가
sns.swarmplot(data=tips, x="total_bill",y ="day", hue="sex")
```

```python
# dodge = True / hue로 나눈 값을 완전히 분리시킬 수 있다. 
#마커가 생략될 수있다.
sns.swarmplot(data=tips, x="total_bill",y ="day", hue="sex", dodge=True)
```

```python
# scatter() 메서드의 키워드 인수 작성으로 커스텀하게 적용 가능
sns.swarmplot(
    data=tips, x="total_bill",y ="day", 
    marker="x", linewidth=1,
)
```

# catplot()
## -> categorical plot = catplot

```python
#axes-level(row,col)까지 분류를 나눌 수 있다.
# kind키워드 인수를 사용하면 그래프 유형을 선택 할 수 있다.
```

```python
sns.catplot(data=df, x="age", y="class", kind="violin", color=".9", inner=None)
sns.swarmplot(data=df, x="age", y="class", size=3)
```

# jointplot()

```python
# 두 개의 변수 분포를 나타낼 때 활용하면 좋은 플롯
# histogram과 scatter plot을 동시에 사용해서 시가걱 효과를 표현
```

```python
# penguins의 DataFrame을 전달
#x = column name인 bill_length_mm / y = column name bill_depth_mm
penguins=sns.load_dataset("penguins")
sns.jointplot(data=penguins, x="bill_length_mm",y="bill_depth_mm")
```

```python
# 어떤 특징을 찾기 어렵다
```

```python
#hug키워드 인수를 활용하여 species column에 대한 변수 추가
#scatterplot에 species별로 색상이 다르게 적용
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
#kind키워드 인수를 하나 추가 ->  kde를 전달
#밀도가 더 눈에 잘 보이도록 변경
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind = "kde")
```

```python
#kind키워드 인수에 reg값을 전달 -> 선형회귀에 대한 결과 
#hue와 kind='reg'는 동시에 사용 불가
adelie_penguins = penguins[penguins["species"] == 'Adelie']
sns.jointplot(data=adelie_penguins, x="bill_length_mm", y="bill_depth_mm", kind = "reg")
```

```python
#hight(크기),ratio(비율),marginal_ticks키워드 인수 추가
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm",hue="species",
              height=7, ratio=2, marginal_ticks=True)
```

# pandas pivot table

```python
#컬럼 중에서 두개의 컬럼을 인덱스,컬럼 로 사용하여 데이터를 조회하여 펼쳐좋은것을 말한다
# 인덱스와 컬럼을 바꿔 인덱스의 라벨값이 첫번째 키의 값과 같고 컬럼의 인덱스의 라벨값이 두번째 키값과 같은 데이터를 찾아서 해당 칸에 넣는다
# 없으면 nan / 마지막 인자 평균값 구하기
```

```python
import pandas as pd
import numpy as np
```

```python
# 컬럼 = who 카테고리 (child,man,woman)
# 인덱스 = class ( 선실 등급)
titanic = sns.load_dataset("titanic")
pd.pivot_table(titanic, "survived", index="class", columns="who")
```

```python
#aggfunc 키워드 인수를 여기선 np.sum으로 전달
pd.pivot_table(titanic, values="survived", index="class",
               columns="who", aggfunc=np.sum)
```

```python
#해당하는 값이 없다면 NaN으로 표현
#pivot_table()메서드 인자중 fill_value에 값을 전달하면 해당 값으로 NaN이 처리
pd.pivot_table(titanic, values="survived", index="class",
               columns="deck",aggfunc=np.mean, fill_value=0)
```

```python
#해당하는 값이 없다면 NaN으로 표현
#pivot_table()메서드 인자중 fill_value에 값을 전달하면 해당 값으로 NaN이 처리
pd.pivot_table(titanic, values="survived", index="class",
               columns="deck",aggfunc=np.mean, fill_value=0)
```

# heatmap()

## 데이터 셋중 비교하고자 하는 column의 pair가 모두 카테고리 값이거나, 피봇테이블의 결과를 가지고 히트맵으로 표현하면 컬럼간의 상관관계를 짙고 옅음으로 표현할 수있다.

```python
#crest테마 적용 = 푸른 빛을 띌수록 생존율이 높다
ttn_cls_who = pd.pivot_table(titanic, "survived", index="class",columns="who")
sns.heatmap(ttn_cls_who, cmap="crest")
```

```python
#annot키워드 인수에 True값 -> cell마다 value표기
#fmt키워드 인수 사용 cell마다 표기한 value에 포맷 지정
# linewidth -> cell간의 구분 명확
sns.heatmap(ttn_cls_who, cmap="mako", annot = True, fmt=".2f", linewidth=.5)
```
